```python
import cv2
import numpy as np
import tensorflow as tf
```

```python
# Load the MobileNet model
model = tf.keras.applications.MobileNetV2()

# Define a function to extract features from a video frame
def extract_features(frame):
  # Resize the frame to the expected input size for MobileNet
  frame = cv2.resize(frame, (224, 224))

  # Preprocess the frame
  frame = frame / 255.0

  # Convert the frame to a NumPy array
  frame = np.array(frame)

  # Expand the frame dimensions to match the expected input shape for MobileNet
  frame = np.expand_dims(frame, axis=0)

  # Extract features from the frame
  features = model.predict(frame)

  # Return the features
  return features

# Extract features from a video
video_path = "datasets/kinetics400_5per/train/abseiling/7dJXS-QdVFc.mp4"

# Open the video file
video = cv2.VideoCapture(video_path)

# Create a list to store the features
features = []

# Loop over the video frames
while video.isOpened():

  # Read the next frame
  ret, frame = video.read()

  # If the frame is not empty, extract features from it
  if ret:
    features.append(extract_features(frame))

  # Otherwise, break the loop
  else:
    break

# Close the video file
video.release()

# Save the features to a file
np.save("features/features-test.npy", features)
```
